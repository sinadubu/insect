```python
from pathlib import Path

# =========================================================
# YOLO 라벨 파일(.txt)의 클래스 ID를 일괄 수정하는 스크립트
# 예: 실수로 '1'로 지정된 클래스를 '0'으로 변경할 때 사용
# =========================================================

# 라벨 파일이 있는 루트 디렉토리 설정
# 주의: 본인의 데이터셋 경로에 맞게 수정해주세요.
label_root = Path(r"C:\dataset\mealworm\labels")

# 하위 폴더를 포함하여 모든 .txt 파일 순회
for txt in label_root.rglob("*.txt"):
    # 파일 내용 읽기 (줄 단위 분리)
    try:
        text = txt.read_text(encoding="utf-8").strip().splitlines()
    except UnicodeDecodeError:
        # 인코딩 에러 발생 시 건너뛰거나 다른 인코딩 시도 가능
        continue
        
    fixed = []
    changed = False
    
    for line in text:
        parts = line.split()
        # YOLO 포맷: <class_id> <x> <y> <w> <h>
        # 첫 번째 항목(class_id)이 '1'인 경우 '0'으로 변경
        if len(parts) > 0 and parts[0] == "1":
            parts[0] = "0"
            changed = True
        
        # 수정된 라인을 다시 조합하여 리스트에 추가
        fixed.append(" ".join(parts))
    
    # 변경사항이 있는 경우에만 파일 덮어쓰기
    if changed:
        txt.write_text("\n".join(fixed), encoding="utf-8")
        print(f"[수정됨] {txt}")
```
