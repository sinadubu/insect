```python
from ultralytics import YOLO
from pathlib import Path

yaml_path = Path(r"C:\dataset\mealworm\data.yaml")

model = YOLO("yolov8s.pt")
results = model.train(
    data=str(yaml_path),
    epochs=80,
    imgsz=640,          # ← 첫 런은 640 권장(안정성). OK면 768로 재런
    batch=16,
    device=0,
    workers=4,
    project="runs",
    name="mealworm_detect_v3",   # 새 폴더로
    exist_ok=True,

    # 학습 안정화/일반화
    cos_lr=True,
    close_mosaic=5,
    rect=False,          # ← 가장 중요 (셔플 가능)
    deterministic=False, # ← 고정 해제
    seed=42,

    # 최적화
    weight_decay=5e-4,
    lr0=0.008,
    lrf=0.01,
    warmup_epochs=3.0,

    # 증강
    hsv_h=0.015, hsv_s=0.6, hsv_v=0.3,
    erasing=0.2,
    mixup=0.0, cutmix=0.0,

    # 성능/로그
    cache=True,          # ← 선택: I/O 가속
    save=True, save_period=5,
)
```
