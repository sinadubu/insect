```python
from ultralytics import YOLO
from pathlib import Path

# 데이터 설정
yaml_path = Path(r"C:\dataset\mealworm\data.yaml")

# YOLO11s 모델 로드
model = YOLO("yolo11s.pt")

model.train(
    data=str(yaml_path),
    epochs=80,
    imgsz=640,          # 입력 이미지 크기 (768x768)

    batch=16,
    device=0,
    workers=4,

    project="runs",
    name="mealworm_y11_detect",


    cos_lr=True,        # 코사인 러닝레이트 스케줄
    close_mosaic=5,     # 마지막 5 epoch에서 mosaic 끄기
    rect=False,         # 랜덤 배치 (셔플 가능 → 일반화↑)
    deterministic=False,# 완전 고정 X (성능↑)
    seed=42,            # 난수 시드


    hsv_v=0.3,          # 기본보다 밝기 변화 폭 키우기 (0.0~1.0 정도에서 조절)


    save=True,
    save_period=5,      # 5 epoch마다 중간 weight 저장
)
```
